```python
#list column name form each of class and its reverse
e_l = ['BFI_1','BFI_11','BFI_16','BFI_26','BFI_36']
e_l_r = ['BFI_6','BFI_21','BFI_31']

a_l = ['BFI_7','BFI_17','BFI_22','BFI_32','BFI_42']
a_l_r = ['BFI_2','BFI_12','BFI_37']

c_l = ['BFI_3','BFI_13','BFI_28','BFI_33','BFI_38']
c_l_r = ['BFI_8','BFI_18','BFI_23', 'BFI_43']

n_l = ['BFI_4','BFI_14','BFI_19','BFI_29','BFI_39']
n_l_r = ['BFI_9','BFI_24','BFI_34']

o_l = ['BFI_5','BFI_10','BFI_15','BFI_20','BFI_25','BFI_30', 'BFI_40', 'BFI_44']
o_l_r = ['BFI_35','BFI_41']
```

```python
#calculate the value for each class
test["Extraversion_Val"] = ((((test[e_l].mean(axis = 1))*5) + (12 - (test[e_l_r].mean(axis = 1))*3))/34)*5
test["Agreeableness_Val"] = ((((test[a_l].mean(axis = 1))*5) + (16 - (test[a_l_r].mean(axis = 1))*4))/37)*5
test["Conscientiousness_Val"] = ((((test[c_l].mean(axis = 1))*5) + (16 - (test[c_l_r].mean(axis = 1))*4))/37)*5
test["Neuroticism_Val"] = ((((test[n_l].mean(axis = 1))*5) + (12 - (test[n_l_r].mean(axis = 1))*3))/34)*5
test["Openness_Val"] = ((((test[o_l].mean(axis = 1))*8) + (8 - (test[o_l_r].mean(axis = 1))*2))/70)*5
```

```python
#Create label by comparing values
columns_to_compare = ["Extraversion_Val", "Agreeableness_Val", "Conscientiousness_Val", "Neuroticism_Val", "Openness_Val"]

# Create a new column 'MaxColumn' with the column name having the maximum value
test['Label'] = test[columns_to_compare].idxmax(axis=1)
test['Label'] = test['Label'].astype('string')
test['Label'] = test['Label'].replace("_Val","", regex=True)
test.to_csv("C:/Users/User/twitter-personality-classification-master/twitter-personality-classification/dataset/label-result-journal.csv")
```
